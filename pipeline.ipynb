## Buscando arquivo

```python
import networkx as nx
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from graphviz import *
```

### Grafo 01

Este grafo contem as interacoes das 102 proteinas de interesse entre si

```python
url102 = 'https://raw.githubusercontent.com/Lezadozo/Tcc/main/102Genes.csv'
df102 = pd.read_csv(url102)
genes102 = []
for x in df102['gene']:
  genes102.append(x)
  
genes102 = '%0d'.join(genes102)
#print(proteins102)
url = 'https://string-db.org/api/tsv/network?identifiers=' + genes102 + '&species=9606'
request_result = requests.get(url)
```

```python
lines = request_result.text.split('\n') 
data = [l.split('\t') for l in lines] 
df = pd.DataFrame(data[1:-1], columns = data[0]) 
print(df.columns)
interactions = df[['preferredName_A', 'preferredName_B', 'score']] 
```

```python
G1=nx.Graph(name='Grafo de interaÃ§Ã£o')
interactions = np.array(interactions)
for i in range(len(interactions)):
    interaction = interactions[i]
    a = interaction[0] # protein a node
    b = interaction[1] # protein b node
    w = float(interaction[2]) # score as weighted edge where high scores = low weight
    G1.add_weighted_edges_from([(a,b,w)]) # add weighted edge to graph 
```

```python
def rescale(l,newmin,newmax):
    arr = list(l)
    return [(x-min(arr))/(max(arr)-min(arr))*(newmax-newmin)+newmin for x in arr]
```

```python
graph_colormap = cm.get_cmap('plasma', 12)
# cor do vertice varia de acordo com o grau
c = rescale([G1.degree(v) for v in G1],0.0,0.9) 
c = [graph_colormap(i) for i in c]
# tamanho do vertice varia de acordo com a centralidade
bc = nx.betweenness_centrality(G1) # betweeness centrality
s =  rescale([v for v in bc.values()], 1000, 10000)
# converter o peso da aresta em forÃ§a de interaÃ§Ã£o (para medida de centralidade)
ew = rescale([float(G1[u][v]['weight']) for u,v in G1.edges], 0.1,4)
# cor das arestas de acordo com seu peso
ec = rescale([float(G1[u][v]['weight']) for u,v in G1.edges],0.1,4)
ec = [graph_colormap(i) for i in ec]
# montar grafo
pos = nx.random_layout(G1)
plt.figure(figsize=(19,9),facecolor=[0.7,0.7,0.7,0.4])
nx.draw_networkx(G1, pos=pos, with_labels=True, node_color=c, node_size=s,edge_color= ec,width=ew,
                 font_color='white',font_weight='bold',font_size='9')
plt.axis('off')
plt.show()
```

### Grafo 02
Este grafo deve conter todas as interaÃ§Ãµes das 102 proteinas de interesse com quaisquer outras proteinas do String-DB

```python
#https://string-db.org/api/[output-format]/network?identifiers=[your_identifiers]&[optional_parameters]

# Este script busca do banco de dados todas as inteaÃ§Ãµes com cada uma das 102 proteinas de interesse
url102 = 'https://raw.githubusercontent.com/Lezadozo/Tcc/main/102Genes.csv'
df102 = pd.read_csv(url102)
genes102 = []
for x in df102['gene']:
  genes102.append(x)  

string_api_url = "https://version-11-5.string-db.org/api"
output_format = "tsv"
method = "network"
params = {
    "identifiers" : "\r".join(genes102), # your protein list
    "species" : 9606, # species NCBI identifier 
    "limit" : 10, # only one (best) identifier per input protein; TODO: How is choosed the best indentifier?
    "echo_query" : 1, # see your input identifiers in the output
    "caller_identity" : "Luan" # your app name

}

request_url = "/".join([string_api_url, output_format, method])
results = requests.post(request_url, data=params)

apiList = []
for line in results.text.strip().split("\n"):
    l = line.split("\t")
    apiList.append([l[2], l[3], l[5]])
    #print(de, para, score, sep="\t")
print(df102)
```

```python
df = pd.DataFrame(apiList[1:-1], columns = ['preferredName_A', 'preferredName_B', 'score']) 
interactions = df[['preferredName_A', 'preferredName_B', 'score']] 

G2=nx.Graph(name='Grafo de interaÃ§Ã£o')
interactions = np.array(interactions)
for i in range(len(interactions)):
    interaction = interactions[i]
    a = interaction[0] # protein a node
    b = interaction[1] # protein b node
    w = float(interaction[2]) # score as weighted edge where high scores = low weight
    G2.add_weighted_edges_from([(a,b,w)]) # add weighted edge to graph
```

```python
graph_colormap = cm.get_cmap('plasma', 12)
# cor do vertice varia de acordo com o grau
c = rescale([G2.degree(v) for v in G2],0.0,0.9) 
c = [graph_colormap(i) for i in c]
# tamanho do vertice varia de acordo com a centralidade
bc = nx.betweenness_centrality(G2) # betweeness centrality
s =  rescale([v for v in bc.values()], 1000, 10000)
# converter o peso da aresta em forÃ§a de interaÃ§Ã£o (para medida de centralidade)
ew = rescale([float(G2[u][v]['weight']) for u,v in G2.edges], 0.1,4)
# cor das arestas de acordo com seu peso
ec = rescale([float(G2[u][v]['weight']) for u,v in G2.edges],0.1,4)
ec = [graph_colormap(i) for i in ec]
# montar grafo
pos = nx.random_layout(G2)
plt.figure(figsize=(19,9),facecolor=[0.7,0.7,0.7,0.4])
nx.draw_networkx(G2, pos=pos, with_labels=True, node_color=c, node_size=s,edge_color= ec,width=ew,
                 font_color='white',font_weight='bold',font_size='9')
plt.axis('off')
plt.show()
```
